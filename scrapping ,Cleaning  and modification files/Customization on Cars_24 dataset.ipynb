```python

Cars_24['Model_Year'] = pd.to_datetime(Cars_24['Model_Year'], format='%Y')

# Extract only the year from the 'Model_Year' column
Cars_24['Model_Year'] = Cars_24['Model_Year'].dt.year
```

```python
Cars_24.dtypes
```

```python
import datetime


current_year = datetime.datetime.now().year


Cars_24['Age_of_Car'] = current_year - Cars_24['Model_Year']
```

```python
Cars_24
```

```python
Cars_24['km Driven_Per_Year'] = Cars_24['Km Driven'] / Cars_24['Age_of_Car']
```

```python
Cars_24['Discounted_Car_Price'].sort_values().unique()	
```

```python

def categorize_Car_Price(i):
    if i < 250000:
        return 'below 2.5 lakh'
    elif i >= 250000 and i < 500000:
        return '2.5 lakh-5 lakh'
    elif i >= 500000 and i < 1000000:
        return '5 lakh-10 lakh'
    elif i >= 1000000 and i < 1500000:
        return '10 lakh-15 lakh'
    else:
        return 'greater than 15 lakh'

Cars_24['Categorize_Car_Price'] = Cars_24['Discounted_Car_Price'].apply(categorize_Car_Price)

```
