```python
import pandas as pd
```

```python
# Reading datasets
Hatchback=pd.read_csv(r"Hatchback.csv")
```

```python
Sedan=pd.read_csv(r"Sedan.csv")
```

```python
SUV=pd.read_csv(r"SUV.csv")
```

```python
#Adding a new column Car_body_Type as we can identify its type
```

```python
Hatchback["Car_Body_Type"]="HatchBack"
```

```python
Sedan["Car_Body_Type"]='Sedan'
```

```python
SUV["Car_Body_Type"]='SUV'
```

```python
#Appending or combining all three dataframe in to single data frame for analysis
```

```python
Cars_24 = pd.concat([Hatchback,Sedan, SUV], ignore_index=True)
```

```python
#dropping unnamed column don't require for analysis
Cars_24.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
Cars_24
```

```python
#Getting Model_Year and Model_Name and Brand from Car_Model
for i in range(len(Cars_24)):
   Cars_24['Model_Year'] = Cars_24['Car_Model'].str[:4]
```

```python
Cars_24['Car_Model'] = Cars_24['Car_Model'].str[5:]
      
```

```python
Cars_24['Brand'] = Cars_24['Car_Model'].str.split(' ').str[0]
```

```python
# next step configuration feature segmented into 4 different columns:-
```

```python
Cars_24[["Km Driven","Ownership","Fuel Type","RTO Office"]]=Cars_24['Configuration'].str.split("\n",expand=True)
```

```python
# Now Configuration is not required for analysis so we drop it
Cars_24.drop('Configuration',axis=1,inplace=True)
```

```python
# next step Discounted_Price feature segmented into 2 different columns:-
```

```python
Cars_24[['Discounted_Car_Price','Baseline Price']]=Cars_24['Discounted_Price'].str.split("\n",expand=True)
```

```python
Cars_24.drop('Discounted_Price',axis=1,inplace=True)
```

```python
# Car transmission :'Manual,Automatic and none'
for i in range(len(Cars_24)):
    if "Manual" in Cars_24['Type'][i]:
        Cars_24.loc[i, 'Car_Type'] = 'Manual'
    elif "Automatic" in Cars_24['Type'][i]:
        Cars_24.loc[i, 'Car_Type'] = 'Automatic'
    else:
        Cars_24.loc[i, 'Car_Type'] = 'None'
 
```

```python
# Replacing Km and ₹ units with "" signifies nothing from dataset
```

```python
Cars_24['Km Driven']=Cars_24['Km Driven'].str.replace("KM","")
```

```python
Cars_24['Discounted_Car_Price']=Cars_24['Discounted_Car_Price'].str.replace("₹","")
```

```python
Cars_24['Baseline Price']=Cars_24['Baseline Price'].str.replace("₹","")
```

```python
Cars_24
```

```python
# replacing ','  with " it signifies nothing from dataset
```

```python
Cars_24['Km Driven']=Cars_24['Km Driven'].str.replace(",","")
Cars_24['Discounted_Car_Price']=Cars_24['Discounted_Car_Price'].str.replace(",","")
```

```python
Cars_24['Baseline Price']=Cars_24['Baseline Price'].str.replace(",","")
```

```python
Cars_24
```
